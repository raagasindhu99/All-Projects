**Kmeans:**
k-means clustering is a method of vector quantization, originally from signal processing, that aims to partition n observations into k clusters in which each observation belongs to the cluster with the nearest mean, serving as a prototype of the cluster.

```python
# Within cluster evaluation
from sklearn.cluster import KMeans        #importing kmeans 
ds1 = KMeans(n_clusters=7)                #as per the given data, it can be seen that there are 7 clusters, hence n=7
ds1.fit(df1.iloc[:,0:3])                  #applying kmeans on dataset
kmeans_pred = ds1.predict(df1.iloc[:,0:3])       #predicting the classes for the data points     
```

```python
kmeans_pred                                      #prediction
```

**Hierarchial Clustering**

```python
from sklearn.cluster import AgglomerativeClustering
hierarchical_cluster = AgglomerativeClustering(n_clusters=7, affinity='euclidean', linkage='ward')
hier_labels = hierarchical_cluster.fit_predict(df1)
```

```python
hier_labels
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.cluster import hierarchy

plt.figure(figsize=(15,10))

linkage_data = linkage(df1, method='ward', metric='euclidean')
dendrogram(linkage_data)
hierarchy.set_link_color_palette(['m', 'c', 'y', 'k'])
#plt.axhline(y=30, color = 'r', linestyle = '--')
plt.show()
```

```python
df1['k_p_cluster'] = kmeans_pred                        #Creating a new column which has predicted classes using Kmeans
df1['k_p_cluster'] = df1['k_p_cluster'] +1           #Adding 1 to the predicted classes as the real classes are in the range of 1-7, reason is to map the classes
                
```

```python
df1['h_p_cluster'] = hier_labels                        #Creating a new column which has predicted  using Hierarchial 
df1['h_p_cluster'] = df1['h_p_cluster'] +1           #Adding 1 to the predicted classes as the real classes are in the range of 1-7, reason is to map the classes
                
```

```python
df1
```

```python
gp_df_k = df1.groupby(['Class','k_p_cluster'],as_index=False)['X1'].count().rename(columns={'X1':'counts'})  #grouping the dataset according to given and Kmeans predicted classes
gp_df_h = df1.groupby(['Class','h_p_cluster'],as_index=False)['X1'].count().rename(columns={'X1':'counts'})  #grouping the dataset according to given and Hier predicted classes
```

```python
gp_df_k
```

```python
gp_df_h
```

 Contingency table
 • TP : The number of data pairs found in the same cluster,
 both in C and in P
 • FP: The number of data pairs found in the same cluster in
 C but in different clusters in P
 • FN : The number of data pairs found in different clusters
 in C but in the same cluster in P
 • TN : The number of data pairs found in different clusters,
 both in C and in P

Basic understanding to create a logic where we can map real class to the predicted class and then calculate scores
 Lets Say a real class - 1 maps with predicted class -7

 1 7 TP (1,7)             
 1 3 FP (1,any-7)    
 2 7 FN (any-1,7)
 2 3 TN (any-1,any-7)

```python
TP_K,FP_K,FN_K,TN_K=0,0,0,0                              #Initialising the values of tp,fp,tn,fn with zeroes
for i in gp_df_k['Class'].unique():                #looping through each real class
  tmp = gp_df_k[gp_df_k.Class==i]                    #filtering the data for that particular real class
  tmp1 = tmp.sort_values(by=['counts'],ascending=[False])      #sorting the above returned df according to counts
  T, P = tmp1.iloc[0,0], tmp1.iloc[0,1]                        #assign the true class and predicted class for the group that has the maximum count 
  TP_K = TP_K+gp_df_k[(gp_df_k.Class==T) & (gp_df_k.k_p_cluster==P)].iloc[0,2] if gp_df_k[(gp_df_k.Class==T) & (gp_df_k.k_p_cluster==P)].shape[0]!=0 else TP_K+0
  FP_K = FP_K+gp_df_k[(gp_df_k.Class==T) & (gp_df_k.k_p_cluster!=P)].iloc[0,2] if gp_df_k[(gp_df_k.Class==T) & (gp_df_k.k_p_cluster!=P)].shape[0]!=0 else FP_K+0
  FN_K = FN_K+gp_df_k[(gp_df_k.Class!=T) & (gp_df_k.k_p_cluster==P)].iloc[0,2] if gp_df_k[(gp_df_k.Class!=T) & (gp_df_k.k_p_cluster==P)].shape[0]!=0 else FN_K+0
  TN_K = TN_K+gp_df_k[(gp_df_k.Class!=T) & (gp_df_k.k_p_cluster!=P)].iloc[0,2] if gp_df_k[(gp_df_k.Class!=T) & (gp_df_k.k_p_cluster!=P)].shape[0]!=0 else TN_K+0
```

```python
TP_H,FP_H,FN_H,TN_H=0,0,0,0                              #Initialising the values of tp,fp,tn,fn with zeroes
for i in gp_df_h['Class'].unique():                #looping through each real class
  tmp = gp_df_h[gp_df_h.Class==i]                    #filtering the data for that particular real class
  tmp1 = tmp.sort_values(by=['counts'],ascending=[False])      #sorting the above returned df according to counts
  T, P = tmp1.iloc[0,0], tmp1.iloc[0,1]                        #assign the true class and predicted class for the group that has the maximum count 
  TP_H = TP_H+gp_df_h[(gp_df_h.Class==T) & (gp_df_h.h_p_cluster==P)].iloc[0,2] if gp_df_h[(gp_df_h.Class==T) & (gp_df_h.h_p_cluster==P)].shape[0]!=0 else TP_H+0
  FP_H = FP_H+gp_df_h[(gp_df_h.Class==T) & (gp_df_h.h_p_cluster!=P)].iloc[0,2] if gp_df_h[(gp_df_h.Class==T) & (gp_df_h.h_p_cluster!=P)].shape[0]!=0 else FP_H+0
  FN_H = FN_H+gp_df_h[(gp_df_h.Class!=T) & (gp_df_h.h_p_cluster==P)].iloc[0,2] if gp_df_h[(gp_df_h.Class!=T) & (gp_df_h.h_p_cluster==P)].shape[0]!=0 else FN_H+0
  TN_H = TN_H+gp_df_h[(gp_df_h.Class!=T) & (gp_df_h.h_p_cluster!=P)].iloc[0,2] if gp_df_h[(gp_df_h.Class!=T) & (gp_df_h.h_p_cluster!=P)].shape[0]!=0 else TN_H+0
```

```python
print("Accuracy Score for Dataset1 using Kmeans = " ,(TP_K+TN_K)/(TP_K+FP_K+FN_K+TN_K))       #caluculating accuracy score for dataset1 with formula
```


Eventhough accuracy score is 1, we cannot confidently say that as mapping classes can be false as they are done based on some random mapping. Hence we have tried to get the maximum accuracy as possible. As we cannot depened on accuracy which is derived using external validation, this model cannot be cosidered.

```python
Precision_K = TP_K/(TP_K+FP_K)                #calculating precision
Recall_K = TP_K/(TP_K+FN_K)                   #calculating recall
```

```python
F1_K = 2 * (Precision_K * Recall_K) / (Precision_K + Recall_K)    #f-1 score is the harmonic mean of precision and recall
```

```python
print("F1 Score for Dataset1 using Kmeans = " ,F1_K)
```

```python
print("Accuracy Score for Dataset1 using Hier = " ,(TP_H+TN_H)/(TP_H+FP_H+FN_H+TN_H))       #caluculating accuracy score for dataset1 with formula
```

```python
Precision_H = TP_H/(TP_H+FP_H)                #calculating precision
Recall_H = TP_H/(TP_H+FN_H)                   #calculating recall
```

```python
F1_H = 2 * (Precision_H * Recall_H) / (Precision_H + Recall_H)    #f-1 score is the harmonic mean of precision and recall
```

```python
print("F1 Score for Dataset1 using Hier = " ,F1_H)
```

**From the above curve we can find elbow at k= 2 and 3. To confirm which is the best k value, we are going to calculate silhoutte score.**

```python
from sklearn.metrics import silhouette_score

sil = []                    #initialising an empty sil list
kmax = 10                   #assuming kmax to be 10

# dissimilarity would not be defined for a single cluster, thus, minimum number of clusters should be 2
for k in range(2, kmax+1):                                     #looping through range 2 to 11
  kmeans = KMeans(n_clusters = k).fit(df2.iloc[:,0:17])         #applying kmeans to dataframe
  labels = kmeans.labels_                                       #generating the predicted classes
  sil.append(silhouette_score(df2.iloc[:,0:17], labels, metric ='euclidean'))            #appending all the silhoutte scores to sil list
```
